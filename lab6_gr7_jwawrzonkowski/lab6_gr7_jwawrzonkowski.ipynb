```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f= lambda x, y : y - x**2
h = 0.2
```

```python
#Euler
x = np.arange(0,3+h,h)
y = [1]*x.size
yEu = [1]*x.size
for i in range(x.size-1):
    yEu[i+1] = y[i] + h*f(x[i], y[i])
    y[i+1] = y[i] + h*(f(x[i],y[i])+f(x[i+1],yEu[i+1]))/2 
```

```python
#RK
yRK = [1]*x.size
for i in range(x.size-1):
    K1 = f(x[i], y[i])
    K2 = f(x[i]+1/2*h, y[i]+1/2*K1*h)
    K3 = f(x[i]+1/2*h, y[i]+1/2*K2*h)
    K4 = f(x[i]+h, y[i]+K3*h)
    yRK[i+1] = yRK[i] + 1/6*(K1+2*K2+2*K3+K4)*h
```

```python
#analytic
xs = np.linspace(0, 3, 1000)
ys = 2+2*xs+xs**2-np.exp(xs)
```

```python
plt.figure(figsize=(20,12))
plt.plot(x, y, "o", ms = 10, label="Eu")
plt.plot(xs, ys, label="analityczne")
plt.plot(x, yRK, "o", ms = 10, label = "RK")

plt.legend()
```
