```python
#dodatkowe
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x, y, z, t : - y - z
g = lambda x, y, z, t : x + 0.1 * y
d = lambda x, y, z, t : 0.1 + z * (x - 14)

jednostki = 50000
czasMax = 500
h = czasMax/jednostki
t = np.arange(0,czasMax,h)
```

```python
#RK
x = [0] * t.size
y = [0] * t.size
z = [0] * t.size


for i in range(t.size-1):
        K1x = f(x[i], y[i], z[i], t[i])
        K1y = g(x[i], y[i], z[i], t[i])
        K1z = d(x[i], y[i], z[i], t[i])
        
        K2x = f(x[i]+1/2*K1x*h, y[i]+1/2*K1y*h, z[i]+1/2*K1z*h, t[i]+1/2*h)
        K2y = g(x[i]+1/2*K1x*h, y[i]+1/2*K1y*h, z[i]+1/2*K1z*h, t[i]+1/2*h)
        K2z = d(x[i]+1/2*K1x*h, y[i]+1/2*K1y*h, z[i]+1/2*K1z*h, t[i]+1/2*h)
        
        K3x = f(x[i]+1/2*K2x*h, y[i]+1/2*K2y*h, z[i]+1/2*K2z*h, t[i]+1/2*h)
        K3y = g(x[i]+1/2*K2x*h, y[i]+1/2*K2y*h, z[i]+1/2*K2z*h, t[i]+1/2*h)
        K3z = d(x[i]+1/2*K2x*h, y[i]+1/2*K2y*h, z[i]+1/2*K2z*h, t[i]+1/2*h)
        
        K4x = f(x[i]+K2x*h, y[i]+K2y*h, z[i]+K2z*h, t[i]+h)
        K4y = g(x[i]+K2x*h, y[i]+K2y*h, z[i]+K2z*h, t[i]+h)
        K4z = d(x[i]+K2x*h, y[i]+K2y*h, z[i]+K2z*h, t[i]+h)
        
        x[i+1] = x[i]+1/6*(K1x+2*K2x+2*K3x+K4x)*h
        y[i+1] = y[i]+1/6*(K1y+2*K2y+2*K3y+K4y)*h
        z[i+1] = z[i]+1/6*(K1z+2*K2z+2*K3z+K4z)*h
        
```

```python

# Plot
ax = plt.figure(figsize=(20,12)).add_subplot(projection='3d')

ax.plot(x,y,z, lw=0.5)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.view_init(10,70)

plt.show()
```
